
**Descrição do dataset**

Dataset ITBI Imposto de Transmissão de Bens Imóveis

A Secretaria de Finanças do município do Recife (autora) e a Empresa Municipal de Informática- EMPREL ( organizadora) têm desenvolvido ao longo dos anos datasets das relações de transações imobiliárias com recolhimento de ITBI, contendo a geolocalização, características dos imóveis e informações sobre as operações.

Há datasets dos anos de 2015 até 2023. Para o presente trabalho, o dataset escolhido foi o do ano de 2023, por ser o mais recente.

Com sua longa série temporal, os datasets são de grande importância, pois possui informações fundamentais para que possamos conhecer os aspectos referentes às transações imobiliárias no município do Recife, bem como como o pagamento do imposto ITBI. É, ainda, um dos principais instrumentos para apoiar a elaboração de políticas públicas de cobrança de tributo - ITBI (âmbito público), como uma otimização do mercado imobiliário.

Sobre o ITBI, é o Imposto sobre Transmissão de Bens Imóveis – ITBI – é um tributo de competência municipal que deve ser pago sempre que ocorre a compra de um imóvel para outro nome. No caso de doação, o imposto a ser pago é o ITCMD para o governo do estado, o que não é o caso em questão.

O ITBI é calculado com base no valor venal do imóvel. Para efeito de apuração da base de cálculo do ITBI, de acordo com a [prefeitura do Recife](https://https://recifeemdia.recife.pe.gov.br/node/926#:~:text=Para%20efeito%20de%20apura%C3%A7%C3%A3o%20da%20base%20de%20c%C3%A1lculo,base%20de%20c%C3%A1lculo%20para%20o%20lan%C3%A7amento%20do%20imposto.):


" O adquirente apresentará sua declaração do valor venal do imóvel e, estando em conformidade com a realidade, consideradas as condições normais de mercado para as transmissões imobiliárias, o valor declarado servirá de base de cálculo para o lançamento do imposto. Se o valor declarado pelo adquirente estiver incompatível com a realidade, a base de cálculo do imposto será avaliada pelo município, em procedimento administrativo próprio. Sobre tal valor será aplicada uma alíquota para se chegar ao total do imposto."

Sobre a alíquota de ITBI, o valor em 2023 foi 3%, podendo variar para 1% nos casos de imóveis financiados pelo Sistema Financeiro de Habitação – SFH e 1,8% conforme casos previstos no artigo 52 da Lei 15.563/91.

##**5.6 Teorema de Bayes**
Calcular a probabilidade do imóvel ser COMERCIAL COM LIXO ORGANICO  sabendo que o mesmo é DE BOA VIAGEM e é uma Loja.

```python
'''
P(lixo_organico) * prob(BOA VIAGEM | COMERCIAL LIXO ORGANICO)
/ P(LIXO ORGANICO) * P(BOA VIAGEM | LIXO ORGANICO) + P(SEM LIXO ORGANICO) * P(BOA VIAGEM | SEM LIXO ORGANICO) + P(RESID) * P(BOA VIAGEM | RESID)
'''
loja_bv = df[(df['bairro'] == 'BOA VIAGEM') & (df['tipo_imovel'] == 'Loja')]


prob_lix_org = len(loja_bv[loja_bv['tipo_ocupa'] == 'COMERCIAL COM LIXO ORGANICO']) / len(loja_bv)

prob_s_lix_org = len(loja_bv[loja_bv['tipo_ocupa'] == 'COMERCIAL SEM LIXO ORGANICO']) / len(loja_bv)

prob_resid = len(loja_bv[loja_bv['tipo_ocupa'] == 'RESIDENCIAL']) / len(loja_bv)

prob_bv_tq_clo = len(loja_bv[(loja_bv['tipo_ocupa'] == 'COMERCIAL COM LIXO ORGANICO') & (loja_bv['bairro'] == 'BOA VIAGEM')]) /len(loja_bv[loja_bv['tipo_ocupa'] == 'COMERCIAL COM LIXO ORGANICO'])

prob_bv_tq_slo = len(loja_bv[(loja_bv['tipo_ocupa'] == 'COMERCIAL SEM LIXO ORGANICO') & (loja_bv['bairro'] == 'BOA VIAGEM')]) /len(loja_bv[loja_bv['tipo_ocupa'] == 'COMERCIAL SEM LIXO ORGANICO'])

prob_final = (prob_lix_org * prob_bv_tq_clo) / ((prob_bv_tq_clo * prob_lix_org) + (prob_s_lix_org * prob_bv_tq_slo))
print(format(prob_final) + ' ou ' '{:.2f}'.format(prob_final * 100) + '%')
```

# **8. Intervalo de Confiança**

```python
#Gráfico Histograma do Intervalo de Confiança 90% das avaliações dos Imóveis Moradia localizados em Boa Viagem
means_bv = [np.mean(df_imov_moradia_bv['valor_avaliacao'].sample(1000)) for _ in range(1000)]

# Calcular a média das médias
mean_of_means = np.mean(means_bv)

# Calcular o intervalo de confiança (95%) usando Z-score
confidence_level = 0.9
z_score = norm.ppf(1 - (1 - confidence_level) / 2)  # z-score for 95% confidence
sample_standard_error = np.std(means_bv, ddof=1) / np.sqrt(len(means_bv))

confidence_interval = (mean_of_means - z_score * sample_standard_error, mean_of_means + z_score * sample_standard_error)

plt.title(f'IC = {confidence_level*100}%')
plt.hist(means_bv, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(confidence_interval[0], color='g', linestyle='dashed', linewidth=1)
plt.axvline(confidence_interval[1], color='g', linestyle='dashed', linewidth=1)
plt.show()

print(confidence_interval)
```
